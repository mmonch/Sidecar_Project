```python
name_description3 = df4[['name', 'description']]
name_description3 = name_description3.rename(columns = {'name': 'Attribute_Business_Name', 'description': 'Attribute_Business_Desc'})
name_description3
```

```python
additional_data = pd.concat([name_description1, name_description2, name_description3]).reset_index().drop(columns =['index'])
additional_data
```

```python
technical_name = []

for i in range(additional_data.shape[0]):
  a = additional_data['Attribute_Business_Name'][i].upper().replace(r" ","_").rstrip("_").replace(r",","").replace(r".","").replace("‘","").replace(r"’","").replace(r"'","")
  technical_name.append(a)
additional_data['Attribute_Technical_Name'] = technical_name
```
